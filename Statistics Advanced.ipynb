```python
# 1. What is a random variable in probability theory?
# Answer:A random variable is a variable in probability theory whose value is not fixed but depends on the outcome of a random experiment.It is a function that assigns a numeric value to each possible outcome of an experiment.
# The value changes randomly every time you repeat the experiment.

# Example:
# Toss a coin:Let X = 1 if Head comes, and 0 if Tail comes.
# Here, X is a random variable.

# Roll a dice:Let Y = the number that appears on the dice.
# Y can be 1,2,3,4,5, or 6 → so Y is a random variable.


# 2.What are the types of random variables?
# Answer:
# Types of Random Variables:
# 1.Discrete Random Variable:
# Takes countable values
# Example: number of heads in 3 coin tosses.

# 2.Continuous Random Variable:
# Takes infinite values in a range
# Example: height of students, time taken to complete a task.


# 3.Explain the difference between discrete and continuous distributions.
# Answer:
# Discrete Distribution:
# A discrete distribution describes variables that take countable, separate values like 0, 1, 2, or 3. Each value has its own probability, and it is represented using a Probability Mass Function (PMF). Examples include dice outcomes or number of students in a class.

# Continuous Distribution:
# A continuous distribution describes variables that can take any value within a range, meaning infinitely many possible values. It uses a Probability Density Function (PDF), and single exact values have zero probability. Examples include height, weight, and time.


# 4.Question 4: What is a binomial distribution, and how is it used in probability?
# Answer:A binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes (success or failure) and a constant probability of success.

# It is used when:
# The experiment is repeated n times,
# Each trial is independent,
# The probability of success p remains the same,
# The outcome of each trial is either success or failure.
# Example:
# If you flip a coin 10 times and want to know the probability of getting exactly 6 heads, you use the binomial distribution.


# 5.What is the standard normal distribution, and why is it important?
# Answer:The standard normal distribution is a bell-shaped curve with mean 0 and standard deviation 1. It is important because it helps us easily calculate probabilities and compare different values using z-scores. Many statistical tests (like Z-test) are based on it, so it makes analysis simpler and more accurate.
# It is important because many statistical methods use it as a reference distribution. By converting any normal variable into a z-score, we can easily calculate probabilities, compare different datasets, and perform hypothesis testing. It makes analysis simpler and universal.


# 6.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
# Answer:The Central Limit Theorem (CLT) says that when you take many samples from any population and calculate their means, those sample means will form a normal (bell-shaped) distribution, even if the original data is not normal—as long as the sample size is large enough.
# The Central Limit Theorem is critical because it allows us to use powerful statistical tools even when the original data is not normal. In real life, most data is messy, skewed, or irregular, but thanks to the CLT, the sample means become normal,


# 7.What is the significance of confidence intervals in statistical analysis?
# Answer:Confidence intervals are important because they give a range of values within which the true population value is likely to lie, rather than giving just a single estimate. This makes the result more reliable and realistic.
# They help us understand the uncertainty in our estimate. For example, saying “the average height is 165 cm” is less useful than saying “the average height is between 162 and 168 cm with 95% confidence.” It shows the possible variation and tells how confident we are in our estimate.


# 8.What is the concept of expected value in a probability distribution?
# Answer:he expected value is the average value you would expect to get if a random experiment is repeated many times. It represents the long-term average outcome of a probability distribution.
# To find it, you multiply each possible value by its probability and then add them up.

# Example:
# For a dice roll,
# Expected value = (1×1/6) + (2×1/6) + … + (6×1/6) = 3.5
# So, the expected value tells you the mean outcome over the long run.


# 9.Write a Python program to generate 1000 random numbers from a normal
# distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
# deviation using NumPy, and draw a histogram to visualize the distribution.
# Answer:
import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(loc=50, scale=5, size=1000)

mean_value = np.mean(data)
std_value = np.std(data)

print("Mean of the data:", mean_value)
print("Standard Deviation of the data:", std_value)

plt.hist(data, bins=30)
plt.title("Histogram of Normal Distribution (Mean=50, SD=5)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# 10.You are working as a data analyst for a retail company. The company has
# collected daily sales data for 2 years and wants you to identify the overall sales trend.
# daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
# 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
#1.Explain how you would apply the Central Limit Theorem to estimate the average sales
# with a 95% confidence interval.
#2. Write the Python code to compute the mean sales and its confidence interval.

# Answer:You have daily sales data. Even if the sales data is not perfectly normal, the Central Limit Theorem (CLT) says:

# When you take repeated samples from the data,

# The sampling distribution of the mean will become approximately normal,

# Even with small datasets.

# So CLT allows us to:
# Estimate the population mean (average sales) using our sample mean.
import numpy as np
from math import sqrt

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
data = np.array(daily_sales)
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)
n = len(data)
z = 1.96

margin = z * (std_sales / sqrt(n))

lower = mean_sales - margin
upper = mean_sales + margin

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", (lower, upper))

```
